```python
# imports
%matplotlib inline
%load_ext autoreload
%autoreload 2
import warnings
warnings.filterwarnings('ignore')
import numpy as np
from loadUniPenData import loadUnipenData
from plotting import plotUniPenData, plotVoronoid
from normalize import normalize_example
from hmmlearn.hmm import GaussianHMM
```

```python
# Load Unipen Data and labels
data, labels = loadUnipenData('../data/pendigits-orig.tra')
# Normalize data within each digit drawing
# And plot an example
# Tdata is the data without PenUp and PenDown
data, tdata = normalize_example(data)
plotUniPenData(data[8])
```

```python
"Classification using Gaussian HMM"
labels = np.array(labels)
models = {}
for i in range(10):
    print "Class", i
    ind = np.where(labels == i)
    digit_data = np.array(data)[ind]
    
    lengths = [len(example) for example in digit_data]
    
    models[i] = GaussianHMM(n_components=10, covariance_type="full",
                          init_params="cmts", params="cmts", verbose=True,
                          tol=50, n_iter=30)
    
    

    models[i].fit(np.atleast_2d(np.concatenate(digit_data)), lengths)
```

```python
test, test_labels = loadUnipenData('pendigits-orig.tes')
test_data, ptest_data = normalize_example(test)

plabels = []
for j in range(len(test_data)):
    llks = np.zeros(10)
    for i in range(10):
        llks[i] = models[i].score(test_data[j])
    plabels.append(np.argmax(llks))


print float(np.sum(np.array(plabels) == np.array(test_labels))) / len(plabels)
```

```python
"Sample From HMM"

import matplotlib.pyplot as plt
from scipy import ndimage


p, axar = plt.subplots(2, 5, figsize=(20, 10))

for i in range(10):
    samples, states =  models[i].sample(10000)

    x = []
    y = []
    for sample in samples:
        x.append(sample[0])
        y.append(sample[1])
    
    heatmap, xedges, yedges = np.histogram2d(x, y, bins=100)
    rt = np.log(ndimage.rotate(heatmap, 90) ** 2)
    extent = [xedges[0], xedges[-1], yedges[0], yedges[-1]]

    #plt.clf()
    axar[i / 5, i % 5].imshow(rt, extent=extent)
```
