```python
# imports
%matplotlib inline
%load_ext autoreload
%autoreload 2
import warnings
warnings.filterwarnings('ignore')
import numpy as np
from loadUniPenData import loadUnipenData
from plotting import plotUniPenData, plotVoronoid
from normalize import normalize_example
from encode import kmeans_encode
from sklearn.cluster import KMeans
from HMMClass import HMMClass
```

```python
# Load Unipen Data and labels
data, labels = loadUnipenData('../data/pendigits-orig.tra')
```

```python
# Normalize data within each digit drawing
# And plot an example
# Tdata is the data without PenUp and PenDown
data, tdata = normalize_example(data)
plotUniPenData(data[8])
```

```python
# Kmeans Encoding
nb_clusters = 254
kmeans = KMeans(init='k-means++', n_clusters=nb_clusters, n_init=10)
kmeans.fit(tdata)
plotVoronoid(tdata, kmeans)
enc_data = kmeans_encode(kmeans, data, nb_clusters)
# Unidimensional example
# PenDown => nb_clusters
# PenUp => nb_clusters + 1
print enc_data[0]
```

```python
'''
Initialization :

Transition matrix is full and all transition are equiprobable
Start probabilites are equiprobable
Emission probabilites are random
'''
centroids = kmeans.cluster_centers_
classifier = HMMClass(10, 254, 10, 50, 30, centroids)
classifier.train(enc_data, labels)
```

```python
# Scoring

test, test_labels = loadUnipenData('pendigits-orig.tes')
test_data, ptest_data = normalize_example(test)
tenc_data = kmeans_encode(kmeans, test_data, nb_clusters)

plabels = classifier.predict(tenc_data)

print float(np.sum(np.array(plabels) == np.array(test_labels))) / len(plabels)
```

```python
'''
Initialization :
Transition probabilites, Emissions probabilities and startprobabilites are
initialized using training data

Kmeans cluster labels are divided in buckets for each hidden states
Start probabilities are 0 except for the hidden state associated with
the PenDown cluster label as each sequence starts with it.

Emission probabilites are derived from the number of occurences of each bucket
in training data

Transition probabilities are derived from the occurences of 2 consecutive
buckets in training data
'''
centroids = kmeans.cluster_centers_
classifier = HMMClass(10, 254, 10, 50, 30, centroids)
classifier.train(enc_data, labels, 'tp')
```

```python
plabels = classifier.predict(tenc_data)

print float(np.sum(np.array(plabels) == np.array(test_labels))) / len(plabels)
```
